```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.discrete.discrete_model import Probit

# Create synthetic data with multicollinearity
np.random.seed(0)
X1 = np.random.rand(100)
X2 = 2 * X1 + np.random.normal(0, 0.1, 100)
y = np.random.randint(0, 2, 100)

# Create a DataFrame for convenience
data = pd.DataFrame({'X1': X1, 'X2': X2, 'y': y})

# Fit a probit regression model with multicollinear independent variables
X_multicollinear = sm.add_constant(data[['X1', 'X2']])
model_multicollinear = Probit(data['y'], X_multicollinear).fit()

# Print the coefficients from the model with multicollinearity
print("Probit Regression with Multicollinearity:")
print(model_multicollinear.summary())

# Slightly change the multicollinear variable (X2)
data['X2'] = 2 * X1 + np.random.normal(0, 0.5, 100)  # Slight change in X2

# Fit a probit regression model with the changed multicollinear variable
X_changed_multicollinear = sm.add_constant(data[['X1', 'X2']])
model_changed_multicollinear = Probit(data['y'], X_changed_multicollinear).fit()

# Print the coefficients from the model with the changed multicollinear variable
print("\nProbit Regression with Slightly Changed Multicollinear Variable:")
print(model_changed_multicollinear.summary())
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.discrete.discrete_model import Probit

# Create synthetic data with multicollinearity
np.random.seed(0)
X1 = np.random.rand(100)
X2 = 2 * X1 + np.random.normal(0, 0.1, 100)
y = np.random.randint(0, 2, 100)

# Create a DataFrame for convenience
data = pd.DataFrame({'X1': X1, 'X2': X2, 'y': y})

# Fit a probit regression model with multicollinear independent variables
X_multicollinear = sm.add_constant(data[['X1', 'X2']])
model_multicollinear = Probit(data['y'], X_multicollinear).fit()

# Print the coefficients from the model with multicollinearity
print("Probit Regression with Multicollinearity:")
print(model_multicollinear.summary())

# Slightly change the multicollinear variable (X2)
data['X2'] = 2 * X1 - np.random.normal(0, 0.1, 100)  # Slight change in X2

# Fit a probit regression model with the changed multicollinear variable
X_changed_multicollinear = sm.add_constant(data[['X1', 'X2']])
model_changed_multicollinear = Probit(data['y'], X_changed_multicollinear).fit()

# Print the coefficients from the model with the changed multicollinear variable
print("\nProbit Regression with Slightly Changed Multicollinear Variable:")
print(model_changed_multicollinear.summary())
```
